```python
new[['STATEFP','COUNTYFP']] = new[['STATEFP','COUNTYFP']].astype(str)
```

```python
new[['CBSA_POP','CBSA_EMP']].describe()
```

```python
new.loc[:, 'STATEFP'] = new['STATEFP'].apply(lambda x: str(x).zfill(2))
new.loc[:, 'COUNTYFP'] = new['COUNTYFP'].apply(lambda x: str(x).zfill(3))
new['COUNTY_NAME'] = new['STATEFP'] + new['COUNTYFP']
```

```python
new
```

```python
my_list = []
with open('fips.txt', 'r') as file:
    for line in file:
        # Process each line as needed
        counties = line.strip().split(maxsplit = 1)
        my_list.append(counties)
```

```python
texas_california_ny = new.loc[(new['STATEFP'] == '48') | (new['STATEFP'] == '06') | (new['STATEFP'] == '36')]
texas_california_ny["FIPS"] = texas_california_ny['STATEFP'] + texas_california_ny['COUNTYFP']
```

```python
# Create a mapping dictionary
mapping_dict = dict(my_list)

# Use map to replace values in 'COUNTY_NAME' column
texas_california_ny['COUNTY_NAME'] = texas_california_ny['COUNTY_NAME'].map(mapping_dict).fillna(texas_california_ny['COUNTY_NAME'])
```

```python


texas_california_ny.dtypes
texas_california_ny
```

```python
texas_california_ny = texas_california_ny.drop(columns = ['COUNTYFP'])
#texas_california_ny['STATE_NAME'] = np.where(texas_california['STATEFP'] == '48', 'Texas', 'California')

texas_california_ny.loc[(texas_california_ny['STATEFP'] == '48', 'STATE_NAME') ]= 'Texas'
texas_california_ny.loc[(texas_california_ny['STATEFP'] == '36', 'STATE_NAME') ]= 'New York'
texas_california_ny.loc[(texas_california_ny['STATEFP'] == '06', 'STATE_NAME') ]= 'California'
texas_california_ny = texas_california_ny.drop(columns = ['STATEFP'])
texas_california_ny
```

```python
grouped = texas_california_ny.groupby(['COUNTY_NAME', 'STATE_NAME', 'FIPS']).mean().reset_index()
new = grouped.loc[grouped['STATE_NAME'] == 'Texas']
```
