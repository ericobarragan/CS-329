# Correlation with Both Datasets

```python
grouped = pd.read_csv('grouped.csv')
grouped.head()
```

```python
grouped.rename(columns={"COUNTY_NAME": "County"}, inplace = True)
grouped.head()
```

```python
merged_df = pd.merge(health, grouped, on='FIPS')
merged_df
```

```python
pd.set_option('display.max_columns',None)
item = merged_df.corr (numeric_only = True)
item

corr = item

thresh = 0.4

mask = corr.abs().gt(thresh)

x = corr.where(mask).stack()
x.to_csv('corr.csv')

item
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

X = merged_df[['NatWalkInd']]
Y = merged_df['% With Access to Exercise Opportunities']

lm.fit(X,Y)
```

```python
print('R^2 = ', lm.score(X,Y))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.regplot(data = merged_df, x = 'NatWalkInd', y = '% With Access to Exercise Opportunities')
plt.ylim(0,)
plt.show()
```

```python
sns.residplot(x= merged_df['NatWalkInd'], y = merged_df['% With Access to Exercise Opportunities'])
plt.show()
```

```python
def PlotPoly(model, x, y, Name):
    x_new = np.linspace(x.min(), x.max(), 1000)
    y_new = model(x_new)

    plt.plot(x, y, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898,0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('NatWalkInd')

    plt.show()
```

```python
x = merged_df['% With Access to Exercise Opportunities']
y = merged_df['NatWalkInd']
f = np.polyfit(x,y,2)
p = np.poly1d (f)
PlotPoly(p,x,y, '% With Access to Exercise Opportunities')
```
